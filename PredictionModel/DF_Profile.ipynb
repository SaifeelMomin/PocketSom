```python
import pandas as pd
import os
```

```python
df_wp = pd.read_csv("winePairing.csv")
df_wp = df_wp.rename({'Unnamed: 0': "Pairing"}, axis = 1 )
df_wp = df_wp.set_index("Pairing")
df_wp.columns = ["category", "beverage", 'food', 'rating']
```

```python
# matched recipes database
df_recipe_wp = pd.read_csv("df_main.csv")
```

```python
wine_profiles = {}
def profiler(wine, b,rf, bf, fl, he, pe, ea, bs, le, ast):
    temp_dict = {}
    total = b+rf+bf+fl+he+pe+ea+bs+le+ast
    wine_profiles[wine] = {'body': b, 'red_fruit': rf, 'black_fruit': bf,
                      'floral':fl, 'herb': he, 'pepper': pe, 'earth': ea, 
                       'baking_sp': bs, 'leather':le, 'astrin':ast, 'total':total}

profiler('Cabernet Sauvignon', 5, 3, 5, 2, 4, 3, 3, 5, 3, 4)
profiler('Merlot', 3,5,3,3,2,1,3,5,2,2)
profiler('Pinot noir', 2,5,0,5,3,0,4,3,1,2)
profiler('Syrah',5,0,5,3,4,5,3,4,3,3)
profiler('Zinfandel', 4,4,4,0,0,5,1,4,2,2)
profiler('Malbec', 4,2,5,3,2,2,3,4,0,2)
profiler('Tempranillo', 4,5,2,0,4,3,4,2,3,4)
profiler('Sangiovese', 4,5,3,0,4,4,3,2,5,4)
profiler('Cabernet Franc', 3,5,0,2,5,2,5,1,0,2)
profiler('Barbera', 3,5,2,1,5,3,5,3,2,4)
profiler('Pinotage', 5,1,5,0,2,3,5,5,0,4)
profiler('Nero', 4,4,4,2,3,4,3,5,3,4)
profiler('Monastrel', 5,0,5,4,5,5,5,5,1,5)
profiler('Petite Sirah', 5,0,5,2,3,2,5,4,0,4)
profiler('Montepulciano', 4,4,2,2,4,4,3,1,3,4)
profiler('Carmenere', 3,5,3,4,5,4,2,3,1,2) 
```

```python
new = df0["nutrition"].str.split(", ", n = 7, expand = True)
df_nutrition = pd.DataFrame(new.values.tolist(), columns=['calories (number)', 'total fat (PDV)', 'sugar (PDV)' , 'sodium (PDV)' , 'protein (PDV)' , 'saturated fat (PDV)', 'carbohydrates (PDV)'])
```

```python
df01 = df0[['category','beverage', 'food', 'rating', 'recipeName', 'ingred', 'ingr_count']]
```

```python
calories = df_nutrition['calories (number)'].values.tolist()
new_cals = []
for cal in calories:
    if type(cal) == float:
        x = 'NA'
        new_cals.append(x)
    else:
        x = float(cal[1:])
        new_cals.append(x)
```

```python
cals = new_cals
fats = df_nutrition['total fat (PDV)'].values.tolist()
sugars = df_nutrition['sugar (PDV)'].values.tolist()
sodiums = df_nutrition['sodium (PDV)'].values.tolist()
prot = df_nutrition['protein (PDV)'].values.tolist() 
sat_fats = df_nutrition['saturated fat (PDV)'].values.tolist()
carbs = df_nutrition['carbohydrates (PDV)'].values.tolist() 
df01['cals'] = cals
df01['fats'] = fats
df01['sugars'] = sugars
df01['sodiums'] = sodiums
df01['proteins'] = prot
df01['sat_fats'] = sat_fats
df01['carbs'] = carbs
```
