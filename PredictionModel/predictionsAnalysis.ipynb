```python
tokenize_wine = tokenize['wine']
predicts_tokens = []
predicts_styles = []

def token_wine(wine):
    for wine in tokenize_wine:
        if  (df['Predicts'].str.contains(wine).any() == True):
            predicts_tokens.append(tokenize['token id'].loc[tokenize['wine'] == wine]), 
            predicts_styles.append(tokenize['style'].loc[tokenize['wine'] == wine])
df['Predicts'].apply(token_wine)
```
