```python
df['protein'] = ((df['protein (pdv)']- df['protein (pdv)'].min())/(df['protein (pdv)'].max() - df['protein (pdv)'].min()) *10).round(2)
df['calories'] = ((df['calories'] - df['calories'].min())/(df['calories'].max() - df['calories'].min()) *10).round(2)
df['fat'] = ((df['fat (pdv)'] - df['fat (pdv)'].min())/(df['fat (pdv)'].max() - df['fat (pdv)'].min()) *10).round(2)
df['sugar'] = ((df['sugar (pdv)'] - df['sugar (pdv)'].min())/(df['sugar (pdv)'].max() - df['sugar (pdv)'].min()) *10).round(2)
df['sodium'] = ((df['sodium (pdv)'] - (df['sodium (pdv)'].min()))/((df['sodium (pdv)'].max()) - (df['sodium (pdv)'].min())) *10).round(2)
df['sat_fat'] = ((df['saturated fat (pdv)'] - df['saturated fat (pdv)'].min())/(df['saturated fat (pdv)'].max() - df['saturated fat (pdv)'].min()) *10).round(2)
df['carbohydrates'] = ((df['carbohydrates (pdv)'] - df['carbohydrates (pdv)'].min())/(df['carbohydrates (pdv)'].max() - df['carbohydrates (pdv)'].min()) *10).round(2)
```

```python
tokenize_wine = tokenize['wine']
predicts_tokens = []
predicts_styles = []

def token_wine(wine):
    for wine in tokenize_wine:
        if  (df['Predicts'].str.contains(wine).any() == True):
            predicts_tokens.append(tokenize['token id'].loc[tokenize['wine'] == wine]), 
            predicts_styles.append(tokenize['style'].loc[tokenize['wine'] == wine])
df['Predicts'].apply(token_wine)
```
