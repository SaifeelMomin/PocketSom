```python
import os
import numpy as np
import pandas as pd
```

```python
path = os.path.join("df_main_3.csv")
df = pd.read_csv(path)
df.columns
```

```python
# Dropping the extra column
df = df.drop(columns='Unnamed: 0')
df
```

```python
df = df.dropna()
df
```

```python
df_wine = df.drop(columns = ['food', 'beverage', 'ind'])
df_wine
```

```python
df["category"].unique()
```

```python
df_wine[['sodium', 'calories', 'fat', 'sodium', 'cholesterol', 'carbs', 'protein']] = df_wine[['sodium', 'calories', 'fat', 'sodium', 'cholesterol', 'carbs', 'protein']].astype(float)
```

```python
df_wine[['Body', 'Tannin', 'Acid', 'ABV', 'Dryness', 'Fruit']] = df_wine[['Body', 'Tannin', 'Acid', 'ABV', 'Dryness', 'Fruit']].astype(int)
```

```python
df["Grape"].unique()
```

```python
df["Color"].unique()
```

```python
df_wine["Color"].replace({"Red ": "Red"}, inplace=True)
```

```python
df_wine['Color'].unique()
```

```python
n_types = df_wine.groupby(['category']).mean().drop(['Body', 'Tannin', 'Acid', 'ABV', 'Dryness', 'Fruit'], axis = 1)
pd.DataFrame(n_types)
```
