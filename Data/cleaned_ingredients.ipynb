```python
import pandas as pd
import pickle as pkl
import os
import numpy as np
from ast import literal_eval
```

```python
with open('ingr_map.pkl', 'rb') as f:
    data = pkl.load(f)
data
```

```python
data = pd.DataFrame(data)
data = data.drop(columns = ['raw_ingr','raw_words','processed','len_proc','count'])
```

```python
data = data.drop_duplicates()
data = dict(zip(data['id'], data['replaced']))
```

```python
recipe_df = pd.read_csv('PocketSom/RAW_recipes.csv')
recipe_df = recipe_df.set_index('id').sort_index()
recipe_df
```

```python
pp_recipe_df = pd.read_csv('PP_recipes.csv')
pp_recipe_df = pp_recipe_df.set_index(['id']).sort_index()
pp_recipe_df
```

```python

recipe_df['ingredient_ids'] = pp_recipe_df['ingredient_ids']
recipe_df = recipe_df[recipe_df['ingredient_ids'].notna()]
recipe_df
```

```python
recipe_df['ingredient_ids'] = recipe_df['ingredient_ids'].apply(literal_eval)
recipe_df['ingredient_ids']
```

```python
recipe_df['clean_ingredients'] = recipe_df.ingredient_ids.apply(lambda x:list(filter(None,map(data.get,x))))
```

```python
clean_df = recipe_df.drop(columns = ['minutes','contributor_id','submitted','n_steps','steps', 'description','ingredients','n_ingredients','ingredient_ids'])

#split and clean 'nutrition' column values, convert to float
clean_df[['calories (number)', 'total fat (PDV)', 'sugar (PDV)' , 'sodium (PDV)' , 'protein (PDV)' , 'saturated fat (PDV)', 'carbohydrates (PDV)']]=clean_df['nutrition'].str.split(', ', expand=True,n=6)
clean_df['calories (number)'] = clean_df['calories (number)'].str.lstrip('[')
clean_df['carbohydrates (PDV)'] = clean_df['carbohydrates (PDV)'].str.strip(']')
clean_df[['calories (number)', 'total fat (PDV)', 'sugar (PDV)' , 'sodium (PDV)' , 'protein (PDV)', 'saturated fat (PDV)', 'carbohydrates (PDV)']] = clean_df[['calories (number)', 'total fat (PDV)', 'sugar (PDV)' , 'sodium (PDV)' , 'protein (PDV)', 'saturated fat (PDV)', 'carbohydrates (PDV)']].astype(float)

#remove very high and very low calorie outliers
clean_df =clean_df[(clean_df["calories (number)"] <800) & (clean_df["calories (number)"] >175) & (clean_df['sugar (PDV)']<50)]
clean_df
```
