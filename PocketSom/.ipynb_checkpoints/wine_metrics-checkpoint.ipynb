```python
import pandas as pd
```

```python
wine_metrics_df = pd.read_csv("wine_metrics.csv", skiprows = 2)
wine_metrics_df = wine_metrics_df.rename(columns = {"Unnamed: 0": "Grape", "Astringent":"Tannin", "Ph": "Acid"})
fruit_intensity = [5,3,5,2,4,5,2,5,2,3,3]
wine_metrics_df["FRUIT"] = fruit_intensity
wine_metrics_df = wine_metrics_df.drop(columns = ["RED FRUIT", "CITRUS FRUIT", "STONE FRUIT", "BLACK FRUIT", "FLORAL", "Minerality", "HERBACIOUS", "PEPPER", "EARTH", "BAKING SPICE", "LEATHER"])
wine_metrics_df = wine_metrics_df.set_index("Grape")
```
