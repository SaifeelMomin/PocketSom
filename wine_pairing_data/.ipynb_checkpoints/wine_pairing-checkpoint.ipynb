```python
import pandas as pd
import numpy as np
```

```python
wine = pd.read_csv("wines.csv")
pairings = pd.read_csv("pairings.csv")
```

```python
wine
```

```python
pairings
```

```python
ingredients = ["chicken", "green beans", "funghi", "oregano", "brie"]
```

```python
dinner = pd.DataFrame()
for x in ingredients:
    for column in pairings.columns:
        dinner = pd.concat([dinner,pairings.loc[pairings[column].isin([x])]])
```

```python
dinner
```

```python
dinner = dinner.groupby("group").mean()
dinner
```

```python
wine_scale = dinner.columns.drop("weight")
```

```python
for column in wine_scale:
    dinner[column] = dinner["weight"] * dinner[column]
dinner
```

```python
dinner.loc['Total',:]= dinner.sum(axis=0)
dinner
```

```python
def get_column_name_for_max_values_of(row):
    return dinner.idxmax(1).loc[row]
```

```python
x = get_column_name_for_max_values_of("Total")
```
